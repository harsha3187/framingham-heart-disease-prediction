<h2>Find Null columns in Data</h2>

```python
PFeature= PFeature.sort_values(by='BAN')
PFeature= PFeature.sort_values(by='BAN')
```

```python
PFeature = PFeature.fillna(0)
TestFeature = TestFeature.fillna(0)
```

```python
df_12 = PFeature.loc[lambda df: ~df.BAN.isin(ban_missing_test)].isin(TestFeature)
```

```python
def ret_non_match_columns(data):
    #print(data.values)
    #print(np.where(~data.values)[0])
    if len(np.where(~data.values)[0])>0:
        return df_12.columns[np.where(~data.values)[0]].tolist()
    else:
        return []
```

```python
df_12['Non_Matching_Columns'] = df_12.apply(lambda df: ret_non_match_columns(df),axis=1)
df_12['Nulls_Present'] = df_12.apply(lambda df: df.values.all(),axis=1)
```

```python
df_nulls = pd.DataFrame()
```

```python
df_nulls['BAN'] = TestFeature['BAN']
df_nulls['Non_Matching_Columns'] = df_12['Non_Matching_Columns']
df_nulls['Nulls_Present'] = df_12['Nulls_Present']
```
